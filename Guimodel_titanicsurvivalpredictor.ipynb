```python
import tkinter as tk
import pandas as pd
import warnings
warnings.filterwarnings('ignore') 
```

```python
model = pd.read_pickle('TitanicSurvival')
```

```python
app=tk.Tk()
app.config(bg='#ECECEC')
app.geometry('479x540')
app.title("Titanic_survival_pred")




tk.Label(app,width=20,text="Titanic Survival Predictor",fg="black",bg="#0078D7",font='Helvetica 15 bold',relief="groove").place(x=90,y=0)

tk.Label(app,width=10,text="Name",bg="#0078D7",fg="black",font="bold",relief="groove").place(x=40,y=80)
tk.Label(app,width=10,text="Age",bg="#0078D7",fg="black",font="bold",relief="groove").place(x=40,y=120)
tk.Label(app,width=10,text="Gender",bg="#0078D7",fg="black",font="bold",relief="groove").place(x=40,y=160)
tk.Label(app,width=15,text="Embarked type",bg="#0078D7",fg="black",font="bold",relief="groove").place(x=240,y=160)
tk.Label(app,width=10,text="Fare",bg="#0078D7",fg="black",font="bold",relief="groove").place(x=40,y=200)
tk.Label(app,width=35,text="Total Siblings/Spouse travelling together",bg="#0078D7",fg="black",font="bold",relief="groove").place(x=40,y=240)
tk.Label(app,width=35,text="Total Parents/Children travelling together",bg="#0078D7",fg="black",font="bold",relief="groove").place(x=40,y=300)
tk.Label(app,width=15,text="Pclass",bg="#0078D7",fg="black",font="bold",relief="groove").place(x=40,y=360)
tk.Label(app,width=20,text="Survival Prediction :",fg="black",bg="#0078D7",font=("Helvetica", 15,"bold"),relief="groove").place(x=40,y=450)

Pclass_types  = [1,2,3]
gender_types = ['female','male']
embarked_types = ['C','Q','S']


name=tk.Variable(app)
age=tk.Variable(app)
pclass = tk.IntVar(app)S
gender = tk.StringVar(app)
embarked = tk.StringVar(app)
fare=tk.Variable(app)
sibsp=tk.Variable(app)
parch=tk.Variable(app)


tk.Entry(app,width=40,textvariable=name,bg='#ffffff',relief="groove").place(x=200,y=80)
tk.Entry(app,width=40,textvariable=age,bg='#ffffff',relief="groove").place(x=200,y=120)
tk.OptionMenu(app,gender, *gender_types).place(x=150,y=160)
tk.OptionMenu(app,embarked, *embarked_types).place(x=390,y=160)
tk.Entry(app,width=40,textvariable=fare,bg='#ffffff',relief="groove").place(x=200,y=200)
tk.Entry(app,width=40,textvariable=sibsp,bg='#ffffff',relief="groove").place(x=40,y=270)
tk.Entry(app,width=40,textvariable=parch,bg='#ffffff',relief="groove").place(x=40,y=330)
tk.OptionMenu(app,pclass,*Pclass_types).place(x=190,y=360)
predict_var = tk.Variable(app)
tk.Label(app,textvariable=predict_var,fg="black",font=("Helvetica", 15,"bold")).place(x=310,y=450)


def predict():
    C=0;Q=0;S=0
    if gender.get()=='male':
        gender_select = 1
    else:
        gender_select = 0
        
    
    if embarked.get()=='C':
        C = 1
    elif embarked.get()=='Q':
        Q = 1
    else:
        S=1
        
    query = pd.DataFrame({'Pclass' : [pclass.get()],'Age' : [age.get()],'SibSp': [sibsp.get()]
                          ,'Parch': [parch.get()],'Fare':[fare.get()],'male': [gender_select],'C': [C],'Q': [Q],'S': [S]})
    predic = model.predict(query)[0]
    if predic==1:
        predict_var.set("Survived")
    else:
        predict_var.set("Dead")
        
        
    
 
tk.Button(app,text="Predict",fg="black",bg="#0078D7",font='arial 15 underline',relief="groove",command=predict).place(x=200,y=400) 
app.mainloop()
```

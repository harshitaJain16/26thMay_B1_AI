```python
x = df.drop(['Placed','ssc_others','etest_p','Mkt&HR','hsc_others','salary','workex','male','Arts','Commerce','Others','Sci&Tech','Science','Comm&Mgmt'],axis=1)
y = df['Placed']
```

```python
x.head()
```

## Decision Tree Classifier

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest =  train_test_split(x,y,test_size=0.25,random_state=200)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python

model_dt = DecisionTreeClassifier().fit(xtrain,ytrain)
```

```python
round(model_dt.score(xtest,ytest)*100,2)
```

```python
imp_dt = pd.DataFrame(model_dt.feature_importances_*100,index=x.columns,columns=['Feature_Importance'])
imp_dt
```

```python
imp_dt.sum()
```

```python
model_rf =  RandomForestClassifier(criterion = 'entropy',min_samples_split=10).fit(xtrain,ytrain)
```

```python
round(model_rf.score(xtest,ytest)*100,2)

```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
ypred = model_dt.predict(xtest)
```

```python
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
```

```python
def missing_age_treatment(x):
    if pd.isnull(x['Age']):
        return df['Age'][(df['Sex']==x['Sex'])&(df['Pclass']==x['Pclass'])].median()
    return x['Age']
```

```python
df['Age'] = df[['Age','Pclass','Sex']].apply(missing_age_treatment, axis=1)
df.info()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.info()
```

```python
df['male'] = pd.get_dummies(df['Sex'],drop_first=True)
```

```python
df.info()
```

```python
df = pd.concat([df, pd.get_dummies(df['Embarked'])],axis=1)
```

```python
df.info()
```

```python
df.drop(['Sex','Embarked'],axis=1, inplace=True)
df.info()
```

```python
x = df.drop('Survived',axis=1)
y = df['Survived']
```

## Splitting the Data

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=200)
```

##  Creating the Machine Model and Training the model with Training Data

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression().fit(xtrain,ytrain)
```

##  Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python

ypred = model.predict(xtest)
```

```python
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
```

```python
round(model.score(xtest,ytest)*100,2)
```

## Deployment

```python
pd.to_pickle(model,'TitanicSurvival')
```
